```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
shape = (1000, 1000, 3)
```

```python
# Red images

red_arr = np.zeros(shape, dtype=np.uint8)
red_arr[:, :, :] = [255,0,0]
# show the image
plt.imshow(red_arr, cmap='gray')
plt.colorbar()  # لإضافة شريط الألوان بجانب الصورة
plt.show()
```

```python
# Green images

green_arr = np.zeros(shape, dtype=np.uint8)
green_arr[:, :, :] = [0, 255, 0]
# show the image
plt.imshow(green_arr, cmap='gray')
plt.colorbar()  # لإضافة شريط الألوان بجانب الصورة
plt.show()
```

```python
# output 1
Red_greenH = np.zeros(shape, dtype=np.uint8)
Red1=red_arr[0:500]
green1=green_arr[500:1000]

Red_greenH = np.concatenate((Red1, green1), axis=0)
# show the image
plt.imshow(Red_greenH, cmap='gray')
plt.colorbar()  # لإضافة شريط الألوان بجانب الصورة
plt.show()
print(Red_greenH.shape)
```

```python
# output 2
Red_greenV = np.zeros(shape, dtype=np.uint8)
Red2=red_arr[0:,0:500]
green2=green_arr[0:,500:1000]

Red_greenV = np.concatenate((Red2,green2), axis=1)
# show the image
plt.imshow(Red_greenV, cmap='gray')
plt.colorbar()  # لإضافة شريط الألوان بجانب الصورة
plt.show()
print(Red_greenV.shape)
```
