```python
#@title Run the following cell code to create data files in directories

import os

def create_nested_folders(base_path, depth):
    current_path = base_path
    for i in range(1, depth + 1):
        current_path = os.path.join(current_path, f'level {i}')
        os.makedirs(current_path, exist_ok=True)
    # Create an empty data.txt file in the last subfolder
    with open(os.path.join(current_path, 'data.txt'), 'w') as f:
        pass
    return os.path.join(current_path, 'data.txt')

# Base directory where the folders will be created
base_dir = 'base'

# Create the base directory if it doesn't exist
os.makedirs(base_dir, exist_ok=True)

list_paths = []

# Create folders with the specified depths, then append to list
list_paths.append(create_nested_folders(os.path.join(base_dir, 'folder 1'), 4))
list_paths.append(create_nested_folders(os.path.join(base_dir, 'folder 2'), 4))
list_paths.append(create_nested_folders(os.path.join(base_dir, 'folder 3'), 3))
list_paths.append(create_nested_folders(os.path.join(base_dir, 'folder 4'), 5))

print("List fof created files in directories:\n")
list_paths
```

```python
#@title run this cell to define move_file(src,dst)

import shutil

def move_file(src, dst):
  """
  Moves a file from the source path to the destination path.

  Args:
    src: The path to the source file.
    dst: The path to the destination directory or file.
  """

  try:
    shutil.move(src, dst)
    print(f"File moved successfully from {src} to {dst}")
  except shutil.Error as e:
    print(f"Error moving file: {e}")
  except FileNotFoundError:
    print(f"Source file not found: {src}")
  except OSError as e:
    print(f"Error moving file: {e}")

print(move_file.__doc__)
```

**Task**: Move each file in the list up one directory

i.e. Move the file from its current location (folder) to the previous folder



image.png

```python
# Solution

# Solution

for src in list_paths:
  dst = "/".join(src.split("/")[:-2]) # add python script to define the previous path of src path
  move_file(src, dst)
```
