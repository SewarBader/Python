**correlation coefficient**



*   The correlation coefficient is an equation that is used to determine

      the strength of the relation between two variables


*   The correlation coefficient always lies between -1 to +1
*    -1 represents X and Y are negatively correlated
*   +1 represents X and Y are positively correlated



---



---


r = n (∑xy) – ∑x ∑y / √ [n* (∑x2 – (∑x)2)] * [n* (∑y2 – (∑y)2)]


*   r = correlation coefficient

*   n = number of observations

*   x = 1st variable in the context



*   y = 2nd variable




---

Absolute Value of r Strength of Relationship


r < 0.3  ==>  None or very weak


0.3 < r <0.5   ==>  Weak


0.5 < r < 0.7  ==>   Moderate


r > 0.7  ==>   Strong

```python


# ////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////


import numpy as np
import math

def get_relation_strength(r:float):

  relation_direction = "Negative"
  if r > 0 : relation_direction = "Positive"
  elif r ==0 : relation_direction = "Nuetral"

  r = math.fabs(r)
  if 0 < r < 0.3 : return f"{relation_direction} None/Very week"
  elif 0.3 >=  r < 0.5 : return f"{relation_direction} Week"
  elif 0.5 < r < 0.7 : return f"{relation_direction} Moderate"
  elif 0.7<= r < 1.0 : return f"{relation_direction} Strong"
  else : return None

def get_corre_coefficient(x: np.array, y: np.array):
    #  r = n (∑xy) – ∑x ∑y / √ [n* (∑x2 – (∑x)2)] * [n* (∑y2 – (∑y)2)]

    if len(x) != len(y):
        print("x array not equal y array")
        return None

    n = len(x)

    sum_x = 0  # ∑x
    sum_y = 0  # ∑y

    sum_xy = 0  # ∑xy

    sum_squared_x = 0  # ∑x2
    sum_squared_y = 0  # ∑y2

    for i in range(n):
        sum_x += x[i]  # ∑x

        sum_y += y[i]  # ∑y

        sum_xy += x[i] * y[i]  # ∑xy

        sum_squared_x += x[i] ** 2  # ∑x2

        sum_squared_y += y[i] ** 2  # ∑y2


    #  r = n (∑xy) – ∑x ∑y / √ [n* (∑x2 – (∑x)2)] * [n* (∑y2 – (∑y)2)]
    r = (n * sum_xy - sum_x * sum_y) / (
        math.sqrt((n * sum_squared_x - (sum_x ** 2)) * (n * sum_squared_y - (sum_y ** 2))))
      # r = ((n * sum_XY) - (sum_X * sum_Y))/((n * sum_X_squared - sum_X ** 2) * (n * sum_Y_squared - sum_Y ** 2)) ** 0.5
    return r

arr_total_roms = df['total_rooms'].values
arr_median_house_value = df["median_house_value"].to_numpy()


corr = get_corre_coefficient(x=arr_total_roms, y=arr_median_house_value)
print("Correlation coefficient : " , corr)
print("strength : ", get_relation_strength(corr))

# x = df["total_rooms"].tolist()
# y = df[["median_house_value"]].values

# correlation = correlation_coefficient(x, y)
# print(f"correlation_coefficient between total_rooms and median_house_value is weak: {correlation}")
```
