Given the following two lists, implement the union, intersection, difference, and symmetric difference operations without using built-in set methods.

```python
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
```

```python
# 2.Find and print the most frequent word in the paragraph.
Paragraph_Filter=re.sub(r'[^-a-zA-Z\s]', '', paragraph).split()
frequent_word = ""
max_frequency = 0

for word in Paragraph_Filter:
    if Paragraph_Filter.count(word)>max_frequency:
        max_frequency = Paragraph_Filter.count(word)
        frequent_word = word
print(f"The most frequent word is '{frequent_word}' with frequency {max_frequency}.")
```

# **Task 3: Remove Redundant Numbers from List**

Given the following list:

```python
nums = [1, 2, 4, 5, 5, 5, 7, 5, 7, 4, 2, 4]
```

1. Remove Redundant Numbers in List Without Using set()

    Write Python code to create a new list that contains the numbers from nums with duplicates removed. Do not use the set() method.

2. Remove Redundant Numbers In-Place (Modify nums List) Without Using set()

    Write Python code to modify the original nums list in-place to remove redundant numbers. Do not use the set() method.

```python
# 1.Remove Redundant Numbers in List Without Using set()

uniqueList=[]
for num in nums:
  if num in uniqueList:
    continue
  else:
      uniqueList.append(num)
print(uniqueList)
```

```python
# 2.Remove Redundant Numbers In-Place (Modify nums List) Without Using set()
nums = [1, 2, 4, 5, 5, 5, 7, 5, 7, 4, 2, 4]
nums.sort()
current=0
for num in range(len(nums)):
 if nums[num] != nums[current]:
        current =current+1
        nums[current] = nums[num]
del nums[current+1 :]
print(nums)
```
